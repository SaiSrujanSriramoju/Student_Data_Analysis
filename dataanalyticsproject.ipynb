```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("student_dataset.csv")
print(df.head())
```


# drop unnamed column


```python
df = df.drop("Unnamed: 0",axis = 1)
print(df.head())
```

# gender distribution

```python
plt.figure(figsize=(5,5))
ax = sns.countplot(data = df,x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
#from the above chart we have analysed that:
#the number of females in the data is more than the number of males
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
sns.heatmap(gb,annot=True)
plt.title("Relationship between Parent's Education and Students Score")
plt.show()
```

```python
#from the above chart we have concluded that the education of the parents have good impact on children
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python
sns.heatmap(gb1,annot=True)
plt.title("Relationship between Parent's Marital Status and Students Score")
plt.show()
```

```python
#from the above chart we have concluded that the marital status of the parents have no major impact on children
```

# Distribution of Ethnic groups

```python
groupA = df.loc[(df["EthnicGroup"]=="group A")].count()
groupB = df.loc[(df["EthnicGroup"]=="group B")].count()
groupC = df.loc[(df["EthnicGroup"]=="group C")].count()
groupD = df.loc[(df["EthnicGroup"]=="group D")].count()
groupE = df.loc[(df["EthnicGroup"]=="group E")].count()

l = ["group A","group B","group C","group D","group E"]
mlist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mlist, labels = l, autopct = "%1.2f%%")
plt.title("Distribution of Ethnic groups")
plt.show()
```

```python
ax = sns.countplot(data = df , x = "EthnicGroup")
ax.bar_label(ax.containers[0])
```

```python
plt.figure(figsize=(8,6))
df.groupby("TestPrep")[["MathScore","ReadingScore","WritingScore"]].mean().plot(kind="bar")
plt.title("Effect of Test Preparation on Scores")
plt.ylabel("Average Score")
plt.show()
```

#Students who completed test prep score significantly higher across all subjects.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.histplot(df["MathScore"], bins=20, kde=True, color="blue", label="Math")
sns.histplot(df["ReadingScore"], bins=20, kde=True, color="green", label="Reading")
sns.histplot(df["WritingScore"], bins=20, kde=True, color="red", label="Writing")
plt.legend()
plt.title("Distribution of Student Scores")
plt.show()

```

#Most students scores cluster around the middle range, with few very high or very low scores.

```python
plt.figure(figsize=(8,6))
df.groupby("Gender")[["MathScore","ReadingScore","WritingScore"]].mean().plot(kind="bar")
plt.title("Average Scores by Gender")
plt.ylabel("Average Score")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="WklyStudyHours", y="MathScore", data=df)
plt.title("Math Scores by Weekly Study Hours")
plt.show()
```
